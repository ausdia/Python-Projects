Create a normalize_docuement function to lowercase, remove white spaces, remove stop words, remove special characters, etc

```python
tokenizer = WordPunctTokenizer()
nltk.download('stopwords')

def normalize_document(inp):

    inp = inp.lower()
    words = tokenizer.tokenize(inp)
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]
    words = [word for word in words if word not in string.punctuation]
    inp = ' '.join(words)
    return inp
```

Calculate document similarity matrix with cosine and Euclidean distances

```python
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(norm_corpus)
```

```python
cosine_similarity_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)
euclidean_distances_matrix = euclidean_distances(tfidf_matrix)
```

```python
print("Cosine Similarity Matrix:")
print(cosine_similarity_matrix)
print("\nEuclidean Distances Matrix:")
print(euclidean_distances_matrix)
```

Cosine:

Documents 1 and 15 have the highest cosine similarity which indicates high similarity. 1 and 6 have medium similarity. The other pairs have scores near 0 which indicates little similarity.

Euclidean:

Documents 1 and 15 have the smallest distance meaning they are very similar. 1 and 6 have a slightly greater distance which means they're decently similar.
